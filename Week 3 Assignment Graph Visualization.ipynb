# Laod the libraries

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
# Set the parameters
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
```

# Read the facebook data

```python
# Read 500 rows
dt = pd.read_table("facebook_combined.txt", sep = " ", names = ['source','target'])
```

```python
dt = dt[300:400]
```

```python
dt.head()
```

```python
dt.shape
```

2.Create basic analysis on the graph, including the graph’s diameter, and at least one other metric of your choosing.  You may either code the functions by hand (to build your intuition and insight), or use functions in an existing package. 

# Make a graph object

```python
G = nx.from_pandas_edgelist(dt)
```

```python
print( nx.info(G))
```

The graph has 94 nodes, 100 edges, average degree is 2.8736

## Diameter`

```python
nx.algorithms.diameter(G)
```

The diameter of the graph is 8.

3.Use a visualization tool of your choice ( NetworkX, Neo4j, Gephi, etc.) to display information.

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
nx.draw(G, node_color='bisque', with_labels=True)
```
